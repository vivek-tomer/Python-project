```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Downloads/Titanic.csv')
```

```python
df
```

```python
# inspection of data
df.info()
```

```python
df["Cabin"] = df["Cabin"].replace(" ","0")
```

```python
# "Totalcharges" has been converted to float type 
df.info()
```

```python
# To check if there is any null values in dataset 
df.isnull()
```

```python
# To check the number of null values in dataset columnwise.
df.isnull().sum()
```

```python
# To check the number of null values in dataset as a whole
df.isnull().sum().sum()
```

```python
# description of a dataset
df.describe()
```

```python
# To check the duplicate in the whole dataset
df.duplicated().sum()
```

```python
# To check the duplicate in the dataset columnwise.
df['Pclass'].duplicated().sum()
```

```python
# To convert the data of "Survived" column from (0 and 1) to (Yes and No).
def conv(value):
    if value == 1:
        return "Yes"
    else:
        return "No"
# Apply this func to the df
df["Survived"] = df["Survived"].apply(conv)
```

```python
df.head()
```

```python
# check the number of people survived with the help of 'survived' column. using seaborn.
plt.figure(figsize= (2,2))     
sns.countplot(x=df['Survived'])
plt.show()
```

```python
# To see the count of passenger's survived.
plt.figure(figsize= (2,3))     
ax = sns.countplot(x=df['Survived'])
ax.bar_label(ax.containers[0])
plt.title("Count of passenger's survived")
plt.show()
```

```python
# To see the number of Survived passenger's in the % form.
plt.figure(figsize= (2,2))                                  # To reduce the chart size.
gb = df.groupby("Survived").agg({"Survived":"count"})
plt.pie(gb['Survived'], labels=gb.index, autopct= "%0.1f%%")    # (labels=gb.index)--> for yes/no in chart, (autopct= "%1.2f%%")--> for auto percentage in chart.
plt.title("percentage of passenger's survived", fontsize= 10)
plt.show()
```

```python
# Exploring by gender first.
plt.figure(figsize= (2,2))  
sns.countplot(x="Sex", data= df, hue= "Survived")
plt.title("Survived by Gender")
plt.show()
```

```python
# according to "Age"
plt.figure(figsize= (30,5))  
sns.countplot(x="Age", data= df, hue= "Survived")
plt.xticks(rotation=90, ha='right')
plt.title("Survived by Age")
plt.show()
```
